# Importing Necessary Libraries for Data Analysis and Visualization









```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

# Loading the Forbes Billionaires 2022 Dataset








```python
df = pd.read_csv("../DataSets/2022_forbes_billionaires.csvnewww.csv")
```

# display the data set

```python
df
```

#  How can we handle missing values in the dataset?
## Checking for Missing Values in the Dataset









```python
df.isnull().sum()
```

### Result : Since all values are 0, it means there are no missing values 

# Cleaning and Converting the Net Worth Column









```python
df.networth = df.networth.replace("[$ B]", "", regex=True).astype(float)
```

### Result: Net Worth Column Cleaned and Converted to Float, where removed $ B symbols



## display the data after cleaning

```python
df
```

# Generate basic statistics such as mean, median, min, max, standard deviation


```python
df.describe()
```

## Result: mean, median, min, max, standard deviation


```python
df.networth.median()
```

# Visualize how billionaire net worth is distributed using bar plots.

```python

plt.figure(figsize=(18,9))
sb.histplot(df.networth,bins=20)
plt.xticks(rotation=90)
plt.show()
```

## Result : Billionaire wealth is highly concentrated, mostly among individuals around age 65 (y= networth,x=counts)

# Visualize how billionaire net worth is distributed using boxplot plots.


```python
plt.figure(figsize=(18,9))
plt.boxplot(df.networth)
plt.xticks(rotation=90)
plt.show()
```

## Result :most billionaires have a net worth below dollar 10 billion, with a few extreme outliers like Elon Musk and Jeff Bezos driving the upper end of the wealth distribution."



















# Identify and display the top 10 billionaires based on net worth using nlargest.

```python
a1=df.groupby('name')['networth'].sum().nlargest(10)
```

```python
a1
```

## Result : like Elon musk, Jeff Bezos ,Bernard Arnault & family  are having largest network 

# Find the top 10 countries with the highest billionaire count using value_counts. 

```python
a=df.country.value_counts().nlargest(10)
```

```python
a
```

## Result :  10 countries with highest billionaire count like United States - 719, China -  515 ,India - 161

# Create a barplot to show the distribution of billionaire wealth across


```python
plt.figure(figsize=(18,9))
sb.barplot(x=df.country, y=df.networth, color='yellow')
plt.xticks(rotation=90)
plt.show()
```

## Result : Most billionaire wealth is concentrated in the United States, with significant contributions from China and India, highlighting geographic disparities in global wealth distribution

# Create a stripplot to show the distribution of billionaire wealth across¶

```python
plt.figure(figsize=(18,9))
sb.stripplot(x=df.country,y=df.networth )
plt.xticks(rotation=90)
plt.show
```

## Result : stripplot is  showing the U.S. leads in billionaire count, while China and India exhibit a wide net worth range with notable outliers y= country , x= networth using stripplot
